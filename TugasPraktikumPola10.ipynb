Estimasi Kepekatan Parameterik dengan distribusi normal

```python
from matplotlib import pyplot
from numpy.random import normal
from numpy import mean
from numpy import std
from scipy.stats import norm
```

```python
# membangkitkan sample 
sample = normal(loc=50, scale=5, size=1000)
```

```python
# menghitung parameters
sample_mean = mean(sample)
sample_std = std(sample)
print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
```

```python
# mendefinisikan distribusi  
dist = norm(sample_mean, sample_std)
values = [value for value in range(30, 70)]
probabilities = [dist.pdf(value) for value in values]
```

```python
# plotting histogram and pdf
pyplot.hist(sample, bins=20, density=True)
pyplot.plot(values, probabilities)
pyplot.show()
```

```python
# membangkitkan sample 
sample = normal(loc=50, scale=5, size=3000)

# plotting histogram and pdf
pyplot.hist(sample, bins=40, density=True)
pyplot.plot(values, probabilities)
pyplot.show()
```

```python
# membangkitkan sample 
sample = normal(loc=50, scale=5, size=5000)

# plotting histogram and pdf
pyplot.hist(sample, bins=60, density=True)
pyplot.plot(values, probabilities)
pyplot.show()
```

Estimasi Kepekatan Parameterik dengan distribusi t

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
```

```python
# Menentukan derajat bebas
degree_of_freedom = 10
```

```python
# Menghasilkan data acak dari distribusi t dengan derajat bebas
data = np.random.standard_t(degree_of_freedom, size=1000)
```

```python
# Membuat histogram dari data
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Menghitung nilai probabilitas dari distribusi t untuk plot kurva
x = np.linspace(min(data), max(data), 1000)
y = t.pdf(x, degree_of_freedom)

# Membuat kurva distribusi t
plt.plot(x, y, 'r', linewidth=2)

plt.xlabel('Nilai')
plt.ylabel('Kepekatan')
plt.title('Distribusi t dengan {} derajat bebas'.format(degree_of_freedom))
```

Estimasi Kepekatan Nonparameterik dengan 3 peak (multimodal)

```python
from matplotlib import pyplot
from numpy.random import normal
from numpy import hstack
from numpy import asarray
from numpy import exp
from sklearn.neighbors import KernelDensity
```

```python
# membangkitkan sample
sample1 = normal(loc=20, scale=5, size=300)
sample2 = normal(loc=40, scale=5, size=700)
sample3 = normal(loc=60, scale=5, size=900)
sample = hstack((sample1, sample2, sample3))
```

```python
# plot histogram
pyplot.hist(sample, bins=50)
pyplot.show()
```

```python
model = KernelDensity(bandwidth=2, kernel='gaussian')
sample = sample.reshape((len(sample), 1))
model.fit(sample)
```

```python
# sample probabilities for a range of outcomes
values = asarray([value for value in range(1, 60)])
values = values.reshape((len(values), 1))
probabilities = model.score_samples(values)
probabilities = exp(probabilities)
```

```python
# plot the histogram and pdf
pyplot.hist(sample, bins=50, density=True)
pyplot.plot(values[:], probabilities)
pyplot.show()
```
