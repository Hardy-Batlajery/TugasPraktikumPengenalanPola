```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#reading the database
data = pd.read_csv("drive/MyDrive/tips.csv")

sns.barplot(x='day', y='tip', data=data, hue='sex')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#reading the database
data = pd.read_csv("drive/MyDrive/tips.csv")

sns.histplot(x='total_bill', data=data, kde=True, hue='sex')

plt.show()
```

```python
# Plotly
import plotly.express as px
import pandas as pd

#reading the database
data = pd.read_csv("drive/MyDrive/tips.csv")

#plotting the scatter chart
fig = px.scatter(data, x="day", y="tip", color="sex")

#showing the plot
fig.show()
```

```python
# Line Chart
import plotly.express as px
import pandas as pd

#reading the database
data = pd.read_csv("drive/MyDrive/tips.csv")

#plotting the scatter chart
fig = px.line(data, y='tip', color='sex')

#showing the plot
fig.show()
```

```python
# Bar Chart
import plotly.express as px
import pandas as pd

#reading the database
data = pd.read_csv("drive/MyDrive/tips.csv")

#plotting the scatter chart
fig = px.bar(data, x = 'day', y ='tip', color = 'sex')

#showing the plot
fig.show()
```

```python
# Histogram
import plotly.express as px
import pandas as pd

#reading the database
data = pd.read_csv("drive/MyDrive/tips.csv")

#plotting scatter the chart
fig = px.histogram(data, x='total_bill', color='sex')

#showing the plot
fig.show()
```
